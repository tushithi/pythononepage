```python
def function_listbuilder(lower_limit, upper_limit):
    lout = []
    if lower_limit <= upper_limit:
        for i in range(lower_limit, upper_limit):
            lout.append(i)
    return lout
        
```

```python
# lets look at the syntax of what we just did
# def <- declares that we are building a function
# function_lisbuilder <- that is the function name
# (): <- follows the function name
# within the curved brackets, there are zero or more arguments, that interact with the function 
# an empty list is created
# there is a check that upper limit is greater than or equal to the lower limit
# currently there is no check that both arguments are integers. 
# looping over the limits set, new items are added to the list
# once the looping is complete, the list is returned
```

```python
# the built function with some arguments is saved to the new varible
built_list = function_listbuilder(0, 3)
```

```python
# lets build a few more functions
def multtwo(i):
    return i*2
def multthree(i):
    return i*3
def plusfour(i):
    return i+4
    
# all pretty simple with a single argument that is modified
```

```python
function_list = {'m2':multtwo, 'm3': multthree, 'p4': plusfour}
# now we have a dictionary of functions 
```

```python
results = {}
for func_name, fx in function_list.items():
    new_list = []
    for item in built_list:
        new_list.append(fx(item))
    results[func_name] = new_list
    
```

```python
for key, values in results.items():
    print(key, values)
```

```python
# we can take the procedure above and turn it into a function as well
def fx_product(b_list, f_list):
    results = {}
    for func_name, fx in f_list.items():
        new_list = []
        for item in b_list:
            new_list.append(fx(item))
        results[func_name] = new_list
    return results
```

```python
final_results = fx_product(built_list, function_list)

# a major thing to notice in this bit here, is the idea of scope. 
# the arguments specified by the function and used inside stay inside, unless fed back out through a return 

# in the function call in this cell, built_list and function_list are fed in. 
# in the function definition, the first and second argument are called differently. 
# but based on the argument order, they are know which is what. 
```

```python
# one could even do this
final_results = fx_product([1, 2, 3, 4], function_list)
```

```python
# but reversing the order would not work, as a list of numbers would be sent to where a list of functions is expected. 
# another way to do this would be 
final_results = fx_product(f_list=function_list, b_list=[1, 2, 3, 4])
# using the key words then directly references the expected arguments. 
# there is a whole variety of ways to do this, mixing up positional and keyword arguments. 
```
